```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
print(pd.__version__)
```

#### we'll study about number of hours used to study and ability to pass the test in Logistic Regression

```python
url = 'https://github.com/prasertcbs/basic-dataset/raw/master/study_hours.csv'
df = pd.read_csv(url)
df.T # transpost and show csv
```

```python
df
```

```python
sns.scatterplot(data=df, x='Hours', y='Pass')
```

```python
sns.lmplot(x='Hours',y='Pass', data=df,
          logistic=False, ci=None, height=4, aspect=1.5,
          line_kws={'color':'orange'})
plt.axhline(.5, color='red', linestyle='--') # just plotting a line at .5
plt.axvline(2.7, color='blue') # or plotting vertical line to diverse between pass and fail still not right

# according to the graph, this data doesn't suitable to use simple linear regression to predict
# instead if the pass was turned to score it may suit lm
```

```python
# df.at[19, 'Hours'] = 100 # trying to change the value
# using the logistic regression
sns.lmplot(x='Hours', y='Pass', data=df,
          logistic=True, height=4, aspect=1.5,
          line_kws={'color':'orange'})
plt.ylabel('Probability of passing exam')
plt.axvline(2.71, color='green', linestyle='--')
plt.axhline(.5, color='red', linestyle='--')

# in other word, logistic line is also called 'sigmoid function'
```

```python
# to prove that the sigmoid function won't go apart 1 we may assign the value to some over rated
df.at[19,'Hours'] = 100
sns.lmplot(x='Hours', y='Pass', data=df,
          logistic=True, height=4, aspect=1.5,
          line_kws={'color':'orange'})
plt.ylabel('Probability of passing exam')
plt.axvline(2.71, color='green', linestyle='--')
plt.axhline(.5, color='red', linestyle='--')

# we could see that even we assign the valur to 100 hours, the probability will not exceed 1
```

# Statsmodels

```python
import statsmodels.api as sm
from patsy import dmatrices # using this to perform logistic regression
```

```python
y, x = dmatrices('Pass ~ Hours', data = df, return_type = 'dataframe') # with intercept column
# y, x = dmatrices('Pass ~ Hours - 1', data = df, return_type = 'dataframe') # without intercept column
```

```python
x.head()
```

```python
y.head()
```

## model_a (with interception)

```python
model_a = sm.Logit(y,x).fit()
model_a.summary()
```

```python
model_a.summary2() # extended model description
```

```python
model_a.predict([[1,2],[1,4]])
```

```python
model_a.predict(x)
```

```python
# predicting the previous model with an adjustment of probability multiplyer
model_a.predict(x).apply(lambda p; 0 if p < .5 else 1)
# lambda is a psudo-function
```
