# Binary Classification: Confusion Matrix

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
print(pd.__version__)
```

```python
# importing data
url = 'https://github.com/prasertcbs/basic-dataset/raw/master/study_hours.csv'
df = pd.read_csv(url)
df.T
```

```python
sns.lmplot(x='Hours', y='Pass', data=df,
          logistic = True, height = 4, aspect = 1.5,
          line_kws={'color':'green'})
plt.ylabel('Probability of Passing the exams')
plt.axvline(2.71, color='orange', linestyle='--')
plt.axhline(.5, color='red', linestyle='--')
```
