```python
import pandas as pd
print(pd.__version__)
```

```python
# data about mammals
df = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/msleep.csv')
df
```

```python
df.info()
# there are many NaN data called 'Missing Data'
```

turning the quality type data to dummy variable so that we could calculate regression effect from the variable

## check NaN in 'vor' column

```python
 df[df.vore.isna()] # show all row that vore is NaN
```

```python
# using a simple dropna function => clean out vore = NaN
df.dropna(subset=['vore'], axis='index', inplace=True)
df
```

## create dummy variable

```python
# turns out value as 1 in the vore type axis
dummies = pd.get_dummies(df['vore'])
dummies
```

```python
dm = pd.concat([df, dummies], axis='columns')
dm
```

# sklearn : linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
dt = dm[['vore', 'bodywt', 'carni', 'herbi', 'omni', 'insecti', 'sleep_total']].dropna(axis='index')
# left carnivore because if all the dummies are 0 then it is carnivore
x = dt[['bodywt','insecti','herbi','omni']]
```

```python
model.fit(x, dt.sleep_total)
```

```python
model.score(x,dt.sleep_total) # R-square
```

```python
model.coef_
```

```python
x
```

# Statmodels: Linear Regression

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
dt.columns
```

```python
model_a = smf.ols(formula='sleep_total ~ bodywt + insecti + herbi + omni', data=dt).fit()
model_a.summary()
```

```python
print(model_a.summary()) # printing brings another format but remain the same value
```

```python
print('intercept :',model.intercept_,'coef :',model.coef_) # sklearn
```

## using c() in statsmodels to automatically create dummy variable internally!

```python
model_b = smf.ols(formula='sleep_total ~ bodywt + C(vore)', data = dt).fit()
model_b.summary()
```
